```python
# Ensure timestamps are datetime
prices_df3['timestamps'] = pd.to_datetime(prices_df['timestamps'])

# Extract year column
prices_df3['year'] = prices_df3['timestamps'].dt.year

# Get the last price of each year
last_prices = prices_df3.groupby(['symbol', 'year'])['price'].last().rename('last_price')

# Shift last prices forward (previous year's last price becomes new year's first price)
first_prices = last_prices.shift(1).rename('first_price')

# Merge first and last prices
returns_df3 = pd.concat([first_prices, last_prices], axis=1).dropna()

# Calculate annual return
returns_df3['annual_return'] = (returns_df3['last_price'] / returns_df3['first_price']) - 1

# Reset index and round results
returns_df3 = returns_df3.reset_index().round(4)

returns_df3
```
