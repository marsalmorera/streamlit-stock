```python
df3.drop(columns='Unnamed: 0', inplace=True)
```

```python
# Change to datetime from stocks. 
df3['timestamps'] = pd.to_datetime(df3['timestamps'])
```

```python
df3['year'] = df3['timestamps'].dt.year
df3['month'] = df3['timestamps'].dt.month
df3['day'] = df3['timestamps'].dt.day
```

```python
cols = ['year', 'month', 'day', 'symbol', 'price']
df3 = df3[cols]
df3.head(5)
```

```python
df3.dtypes
```

```python
# Assuming your DataFrame is named df3
# Filter for January 1st
df_jan1 = df3[(df3['month'] == 1) & (df3['day'] == 1)]

# Sort by symbol and year to ensure proper order
df_jan1 = df_jan1.sort_values(by=['symbol', 'year']).reset_index(drop=True)

# Calculate annual return using lambda function
annual_returns = df_jan1.groupby('symbol').apply(
    lambda group: pd.Series(
        (group['price'].values[1:] - group['price'].values[:-1]) / group['price'].values[:-1],
        index=group['year'].values[1:]
    )
).reset_index(name='annual_return')
```

```python
annual_returns
```

```python
annual_returns = annual_returns.rename(columns={'level_1': 'year'})
```

```python
pivoted_df = annual_returns.pivot(index='symbol', columns='year', values='annual_return')
```

```python
pivoted_df = pivoted_df.loc[:, pivoted_df.columns >= 2000]
pivoted_df
```

```python
# Strip any leading/trailing spaces from column names
df3.columns = df3.columns.str.strip()

# Check column names again
print(df3.columns)
```
